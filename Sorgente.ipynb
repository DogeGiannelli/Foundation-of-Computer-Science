### 1. Normalize the location field of *Users* dataset, splitting into city, region, country.

```python
city_list = list()
territory_list = list()
state_list = list()
for location in users['Location']: #iterazione nella colonna Location
  location_list = location.split(',') #Splitto in base alla virgola e creo lista
  location_list=[location_list[i].strip() for i in range(0,len(location_list))]#tolgo spazi inizio e fine stringa
 
  if len(location_list)==1:
    city_list.append(location_list[0])
    territory_list.append(' ')
    state_list.append(' ')
  elif len(location_list)==2:
    city_list.append(location_list[0])
    territory_list.append(location_list[1])
    state_list.append(' ')
  elif len(location_list)==3:
    city_list.append(location_list[0])
    territory_list.append(location_list[1])
    state_list.append(location_list[2]) 
  else:
    city_list.append(location_list[0])
    territory_list.append(location_list[1])
    state_list.append(location_list[2]) 
    
users['City']=city_list
users['Territory']=territory_list
users['State']=state_list
```

### 5. For each book in `merged books` compute its average rating.

The average is computed considering all books in `books` that have been merged.

```python
#per calcolare l'avarage rating necessitavamo di tutte le riedizioni di ogni libro, cioè degli ISBN

books2 = books.groupby(['Book-Title', 'Book-Author', 'Publisher']).count() #Effettuo un groupby per raggruppare i libri che presentano stesso titolo, autore e casa editrice
merged_books = books2[books2['ISBN']>=2] #prendo in considerazione solo i libri che presentano almeno due volte la stessa tupla
merged_books = pd.merge(books, merged_books, on=['Book-Title', 'Book-Author', 'Publisher'])#effettuo un merge tra le due tabelle books e merged_books
merged_books = merged_books.drop(['Year-Of-Publication_y','Image-URL-S_y', 'Image-URL-M_y', 'Image-URL-L_y', 'ISBN_y'], axis =1)#elimino le colonne che contengono i valori del precedente count
merged_book = merged_books.rename(columns = {'ISBN_x':'ISBN', 'Year-Of-Publication_x':'Year-Of-Publication','Image-URL-S_x':'Image-URL-S','Image-URL-M_x':'Image-URL-M','Image-URL-L_x':'Image-URL-L'})#aggiungo alle colonne già presenti in book la colonna "righe_condivise" dimostrando che i libri presentano la tupla almeno due volte
merged_book
```

```python
#Media calcolata in base al titolo
rat_merged_books = pd.merge(merged_book, book_ratings, on = 'ISBN', how = 'left')
pd.DataFrame(rat_merged_books.groupby(['Book-Title']).mean()[['Book-Rating']])

#OPPURE abbiamo calcolato la media del rating condizionatamente alla pubblicazione

rat_merged_books = pd.merge(merged_book, book_ratings, on = 'ISBN', how = 'left')
sub = pd.DataFrame(rat_merged_books.groupby(['Book-Title','ISBN']).mean()[['Book-Rating']]).rename(columns={'Book-Rating':'Book_Rating_AVG'})
sub.reset_index(inplace = True)
sub
```

### 7. For each book in `goodbooks`, compute the list of its authors. Assuming that the number of reviews with a text (column `work_text_reviews_count`) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the *shared number of reviews with a text*.

```python
diz_autori={}
lista_numero_autori=[]
for i in range(0,99):
  diz_autori[good_books.loc[i, 'original_title']]=good_books.loc[i, 'authors']  #creo un dizionario in cui la chiave è il titolo del libro, il valore associato alla chiave sono gli autori che hanno scritto quel determinato libro
  lista_autori=good_books.loc[i, 'authors'].split(',')
  lista_numero_autori.append(len(lista_autori))  #creo una lista che serve nel punto dopo --> numero di autori che hanno scritto un determinato libro
#diz_autori
```

```python
good_books['numero_autori'] = lista_numero_autori #lista con numero di autori 
#Supponiamo che nel caso in cui ci siano più autori, le reviwes siano ugualmente distrubuite tra gli autori
good_books['shared_number_reviews_book'] = good_books['work_text_reviews_count'] // good_books['numero_autori'] #perciò divido il numero di recensioni per il numero di autori che hanno scritto un libro
good_books.head()
```

```python
shared_dataset = good_books[['authors', 'original_publication_year', 'shared_number_reviews_book', 'numero_autori']]  #per calcolare le shared number of reviews with a text considero queste colonne di interesse

for i in list(range(0,99)):  
  if shared_dataset.loc[i, 'numero_autori'] > 1:  # applico un for e per ogni riga di "shared_dataset" valuto se c'è più di un autore
    lista_split_autori = shared_dataset.loc[i, 'authors'].split(',')  # creo una lista che contiene gli autori 
    for autore in lista_split_autori:  # itero per ogni elemento della lista
      new_riga = [(autore, shared_dataset.loc[i, 'original_publication_year'], shared_dataset.loc[i, 'shared_number_reviews_book'], 1)]  
      # creo una nuova riga che contiene un singolo autore della lista, a cui è associato il corrispondente "original_publication_year" e "shared_number_reviews_book"; alla colonna "numero_autori" associo il valore 1
      dfNew = pd.DataFrame(new_riga, columns = ['authors', 'original_publication_year', 'shared_number_reviews_book', 'numero_autori'])  # aggiungo tale riga a un nuovo dataset
      shared_dataset=shared_dataset.append(dfNew,ignore_index=True)  # metto insieme i due dataset tramite append
shared_dataset = shared_dataset[shared_dataset['numero_autori'] == 1]  # considero solo le righe che hanno 1 come valore su "numero_autori"
del shared_dataset['numero_autori']  # elimino la colonna "numero_autori"
shared_dataset = shared_dataset.rename(columns={'authors':'autore'})
shared_dataset
```

```python
shared_totale = shared_dataset.groupby(['autore'], as_index=False)[['shared_number_reviews_book']].sum() #calcolo la somma totale delle "shared_number_reviews_book" raggruppando per autore 
shared_totale = shared_totale.rename(columns={'shared_number_reviews_book':'shared_number_reviews_with_text'}) #ora ho le shared number of reviews with a text per ogni autore
shared_totale
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

```python
shared_dataset['original_publication_year'] = shared_dataset['original_publication_year'].astype(int) #trasformo l'anno di pubblicazione da float a intero, per una migliore comprensione del dato
somma_shared_reviews = shared_dataset.groupby(['original_publication_year', 'autore'], as_index=False)[['shared_number_reviews_book']].sum() 
#sommo le shared reviews raggruppando per anno di pubblicazione e autore, nel caso in cui un autore abbia pubblicato più libri nello stesso anno
somma_shared_reviews = somma_shared_reviews.rename(columns={'shared_number_reviews_book':'shared_number_reviews_author_year'})
somma_shared_reviews
```

```python
#Creo il dataset in cui vi è l'anno ed il corrispondente numero massimo di recensioni
max_df=pd.DataFrame(somma_shared_reviews.groupby('original_publication_year')['shared_number_reviews_author_year'].max())
max_df

shared_max_year=pd.DataFrame()
for max in max_df['shared_number_reviews_author_year']:#iterazione entro la colonna 'numero di recensioni' del dataset appena creato
  temp = somma_shared_reviews[somma_shared_reviews['shared_number_reviews_author_year'] == max]#Estraggo tutte le righe che hanno valore massimo, in modo tale da avere anche le righe che riportano lo stesso valore massimo 
  shared_max_year = shared_max_year.append(temp, ignore_index = True)

shared_max_year = shared_max_year.rename(columns={'shared_number_reviews_author_year':'max_shared_number_reviews_year'})
shared_max_year.head(50)#Compaiono più autori per lo stesso anno nel caso in cui hanno avuto recensioni condivise e la frequenza di queste ultime corrisponde alla massima
```

```python
both_ratings = pd.merge(books_both_datasets, book_ratings[['ISBN', 'Book-Rating']], on='ISBN') 
#faccio un merge tra il dataset appena creato e 'book_ratings', per avere in 'both_ratings' i ratings che fanno riferimento a 'books'. Per i good_books ho già una colonna che esprime l'average rating
both_ratings = both_ratings.rename(columns={'average_rating':'Avarege_rating_good_books', 'Book-Rating':'Rating_books'})
both_ratings.head()
```

```python
mean_book_rating = both_ratings.groupby('ISBN').mean()['Rating_books'] #media dei ratings che fanno riferimento ai books, raggruppati per 'ISBN' che identifica il libro
mean_goodbooks_rating = both_ratings.groupby('ISBN').mean()['Avarege_rating_good_books'] #stesso calcolo di prima: tale calcolo serve esclusivamente per condensare in una riga il valore dell'average rating per i good_books 
mean_both_rating = pd.merge(mean_book_rating, mean_goodbooks_rating, on='ISBN') #mergiamo i due dataset appena creati, che avranno l'ISBN e le due medie come colonne 
mean_both_rating = mean_both_rating.rename(columns={'Rating_books':'Avarege_Rating_books'})
mean_both_rating['rating_difference'] = mean_both_rating['Avarege_Rating_books'] - mean_both_rating['Avarege_rating_good_books'] #calcoliamo la differenza delle due medie
mean_both_rating
```
